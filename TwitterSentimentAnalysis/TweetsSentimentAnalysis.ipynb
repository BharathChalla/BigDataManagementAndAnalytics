```
#parameters

#input file path
inputPath = "s3://cs6350-bigdata-fall2022/Tweets.csv"

#output file path 
outputPath = ""
```

```
#Step 1

#Loading the file
data = spark.read.option("header","true").option("inferschema","true").csv(inputPath)
display(data)
```

```
#Step 1
#Drop rows where the text field is null

data = data.na.drop(subset=["text"])
display(data)
```

```
#Splitting training and testing dataset

training, test = data.randomSplit([0.8, 0.2])
```

```
#training
```

```
#Step2 - Preprocessing

from pyspark.ml.feature import HashingTF, Tokenizer, StopWordsRemover, StringIndexer
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline

#Tokenizer - convert the text column by braeaking into words
tokenizer = Tokenizer(inputCol = "text", outputCol = "words")

#StopWordRemover - Remove stop-words
remover = StopWordsRemover(inputCol = tokenizer.getOutputCol(), outputCol = "filtered")

#TermHashing - Convert words to term-frequency vectors
hashingTF = HashingTF(inputCol = remover.getOutputCol(), outputCol = "features")

#Label Connversion
indexer = StringIndexer(inputCol="airline_sentiment", outputCol="categoryIndex")

#Logistic Regression
lr = LogisticRegression(featuresCol='features', labelCol='categoryIndex', maxIter=5)

#Creating pipeline
pipeline = Pipeline(stages=[tokenizer, remover, hashingTF, indexer, lr])

```

```
#parameter tuning
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder

paramGrid = ParamGridBuilder() \
                .addGrid(hashingTF.numFeatures, [10, 100, 1000]) \
                .addGrid(lr.regParam, [0.1, 0.01]) \
                .build()
```

```
paramGrid
```

```
#CrossValidator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

crossval = CrossValidator(estimator=pipeline,
                          estimatorParamMaps=paramGrid,
                          evaluator=MulticlassClassificationEvaluator(labelCol ="categoryIndex", predictionCol ="prediction"),
                          numFolds=3) 
```

```
#Train the model
cvModel = crossval.fit(training)
```

```
cvModel.avgMetrics 
```

```
best_params = zip(cvModel.avgMetrics, cvModel.getEstimatorParamMaps())
best_param = max(best_params, key=lambda tup: tup[0])
print(best_param[0])
for k, v in best_param[1].items():
    print(k.name, v)
# params = [lr.regParam.name, hashingTF.numFeatures.name]
# for stage in cvModel.bestModel.stages:
#     print ('stages: {}'.format(stage))
#     print (stage.params, '\n')
# print(cvModel.bestModel.stages[2].getNumFeatures())
# print(cvModel.bestModel.stages[4].getRegParam())
```

```
cvModel.bestModel.
```

```
#Test the model
prediction = cvModel.transform(test)
#display(prediction)
```

```
#Converting results of the test dataframe to RDD
predictionRDD = prediction.select("prediction", "categoryIndex").rdd
#predictionRDD.collect()

#Compute raw scores on the dataset
predictionAndLabels = predictionRDD.map(lambda lp: (float(lp[0]), float(lp[1])))
#predictionAndLabels.collect()
```

```
from pyspark.mllib.evaluation import MulticlassMetrics


# Instantiate metrics object
metrics = MulticlassMetrics(predictionAndLabels)

#Overall Accuracy
accuracy = metrics.accuracy
print("Overall accuracy = " + str(accuracy))

#Confusion matrix
confusionMatrix = metrics.confusionMatrix().toArray()
type(confusionMatrix)
print("Confusion Matrix = " + str(confusionMatrix))

 
# Statistics by class
labels = predictionRDD.map(lambda lp: lp.categoryIndex).distinct().collect()
for label in sorted(labels):
    print("Class %s precision = %s" % (label, metrics.precision(label)))
    print("Class %s recall = %s" % (label, metrics.recall(label)))
    print("Class %s F1 Measure = %s" % (label, metrics.fMeasure(label, beta=1.0)))

# Weighted stats
print("Weighted recall = %s" % metrics.weightedRecall)
print("Weighted precision = %s" % metrics.weightedPrecision)
print("Weighted F(1) Score = %s" % metrics.weightedFMeasure())
print("Weighted F(0.5) Score = %s" % metrics.weightedFMeasure(beta=0.5))
print("Weighted false positive rate = %s" % metrics.weightedFalsePositiveRate)
```

```
#Storing the results in the file.

results = "--------------Evaluation metrics output for the Logistic regression model---------------------"
results += "\n Overall accuracy = " + str(accuracy)
results += "\n Confusion Matrix = " + str(confusionMatrix)

results += "\n-------- Statics by class--------"
for label in sorted(labels):
  results += "\n --- Label:" + str(label) + "------------"
  results += "\n Precision = " + str(metrics.precision(label))
  results += "\n Recall = " + str(metrics.recall(label))
  results += "\n F1 measure = " + str(metrics.fMeasure(label, beta=1.0))
    
results += "\n-------- Weighed stats-------\n"
 
results += "\nWeighted recall  = " + str(metrics.weightedRecall)
results += "\n Weighted precision = " + str(metrics.weightedPrecision)
results += "\n Weighted F(1) Score = " + str(metrics.weightedFMeasure())
results += "\n Weighted F(0.5) Score = " + str(metrics.weightedFMeasure(beta=0.5))
results += "\n Weighted false positive rate  = " + str(metrics.weightedFalsePositiveRate)
```
